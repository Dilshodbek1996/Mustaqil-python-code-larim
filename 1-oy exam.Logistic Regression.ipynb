# Handling with Missing values

```python
# Finding Missing values

missing_columns=df.isnull().sum()
missing_columns=missing_columns[missing_columns>0]

if not missing_columns.empty:
    print((f"MISSING VALUES:\n{missing_columns}"))
else:
    print("No missining values")
```

```python
# Missing values larni umumiy formula bilan to'ldirib ketish

missing_columns=df.columns[df.isnull().sum()>0]

for col in missing_columns:
    if df[col].dtype=="object": 
        df[col].fillna(df[col].mode()[0],inplace=True) # object tipidagi NaN qiymatlarni mode()(eng ko'p uchragan qiymatni tanlab olish) funksiyasi yordamida to'ldirish buyrug'i
    else:
        df[col].fillna(df[col].mean(),inplace=True) # int va float tipidagilarni mean()( o'rtacha qiymat) funfsiyasi bilan to'ldirish
```

```python
# Missing values larni qaytadan tekshiramiz

missing_columns=df.isnull().sum()
missing_columns=missing_columns[missing_columns>0]

if not missing_columns.empty:
    print((f"MISSING VALUES:\n{missing_columns}"))
else:
    print("No missining values")
```

```python
#Mapping

df["Burnout_Symptoms_binary"]=df["Burnout_Symptoms"].map({
    "Occasional":0,
    "No":1,
    "3":1,
    "Yes":2
})
```

# Encoding (from categorical to numerical)

```python
# Finding cardinality

cardinality=df[categorical_col].nunique()
print(cardinality)
```

```python
# Encoding (One_hot encoding) 
# with get_dummies ( cardinality o'rtacha 5 atrofida chiqqani (cardinality low) uchun)

one_hot_encoding=pd.get_dummies(df,columns=categorical_col,drop_first=True) # pandas kutubxonasi yordamida one-hot encoding turidan foydalandim
one_hot_encoding
```

```python
df=one_hot_encoding
df.info()
```

```python
df[df.columns]=df[df.columns].astype(int) # Qo'shilgan yangi ustunlarni booleandan int ga o'tkazish
df.info()
```

# Scaling

# Training Process

```python
# Selecting Target(Output) and Features(Input)

x=df.drop("Burnout_Symptoms_binary",axis=1) # Features (Input)
y=df["Burnout_Symptoms_binary"] # target (Output)
```

```python
y.head()
```

```python
x.head()
```

```python
# Data splitting

x_train,x_temp,y_train,y_temp=train_test_split(x,y, test_size=0.3, random_state=42) # 70 % datasetni traing uchun belgiladik 

x_test, x_val, y_test, y_val=train_test_split(x_temp,y_temp, test_size=0.5, random_state=42) # 30% datasetdan 15 % ini validating uchun qolgan 15 foizi testing uchun
```

```python
print(x_train.shape)
print(x_val.shape)
print(x_test.shape)

print(y_test.shape)
print(y_val.shape)
```

```python
y_train[:10]
```

```python
model=LogisticRegression(multi_class="multinomial",solver="lbfgs",max_iter=200) # selecting model

model.fit(x_train,y_train) # teaching model 
```

```python
y.head() # old features
```

```python
y_pred=model.predict(x_test)# predicting 
print(f"Predict columns:\n{y_pred[:5]}")
```

```python
# Accurary score

accuracy=accuracy_score(y_test,y_pred)
print(f"Accurary:{accuracy*100:.2f}%")
```

```python
# Classification report (Prediction,Recall,F1-score)
print("Classification Report:")
print(classification_report(y_test,y_pred))
```
