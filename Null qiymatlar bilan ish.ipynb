```python
missing_col=df.columns[df.isnull().sum()>0]

for col in missing_col:
    if df[col].dtypes=="object":
        df[col].fillna(df[col].mode()[0],inplace=True)
    else:
        df[col].fillna(df[col].mean(),inplace=True)
```

```python
categorical_col=df.select_dtypes(include=["object","category"]).columns.tolist()
categorical_col
```

```python
one_hot_encoder=pd.get_dummies(df ,columns=categorical_col,drop_first=True)
one_hot_encoder
```

```python
one_hot_encoder.info()
```

```python
one_hot_encoder.astype(int)
```

```python
df=one_hot_encoder.astype(int)
```

```python
df.info()
```

```python
# Label encoding bilan o'zgartirish

from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()

for col in categorical_col:
    df[col]=label_encoder.fit_transform(df[col])
```
