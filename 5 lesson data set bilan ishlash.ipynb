```python
# Qaysi ustun string ya'ni object ligini  aniqlab olish
categorical_col=df.select_dtypes(include=['object','category']).columns
categorical_col
```

```python
# Categorical(string) dan Numerical (Int)ga o'tish
# one_hot encoding

df_one_hot=pd.get_dummies(df,columns=categorical_col,drop_first=True)
```

```python
df_one_hot.head(10)
```

```python
df_one_hot.astype(int)
```

```python
df_one_hot.info()
```

```python
#Categorical(string) dan Numerical (Int)ga o'tish
# Label encoding

numeric_columns = df_one_hot.select_dtypes(include=['float64', 'int64']).columns
```

```python
df_one_hot[numeric_columns] = df_one_hot[numeric_columns].fillna(df_one_hot[numeric_columns].mean())
df_one_hot
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder= LabelEncoder()
for col in categorical_col:
    df[col]=label_encoder.fit_transform(df[col])
df.head(20)
```

```python
#Visual ko'rinish
numerical_col=df.select_dtypes(include=['int64', 'float64']).columns
df[numerical_col].hist(bins=20, figsize=(15,20),color='blue', edgecolor='red')
plt.suptitle('Histogram', fontsize=16)
plt.show()
```

```python
#Bar chart 
categorical_col=df.select_dtypes(include=['object','category']).columns
for col in categorical_col:
    df[col].value_counts().plot(kind='bar', figsize=(8,4),color='green', edgecolor='black')
    plt.title('Bar plot', fontsize=14)
    plt.ylabel('count')
    plt.xlabel(col)
    plt.show()
```
